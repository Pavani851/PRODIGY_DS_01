```python

import zipfile
import os

# Replace with your uploaded filename
zip_path = "API_SP.POP.TOTL_DS2_en_csv_v2_390854.zip"
extract_path = "extracted_files"

# Extract it
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# List the files to find your CSV
os.listdir(extract_path)
```

```python

import pandas as pd

# Path to your main data CSV
csv_path = f"{extract_path}/API_SP.POP.TOTL_DS2_en_csv_v2_390854.csv"

# Skip metadata rows (first 4 rows)
df = pd.read_csv(csv_path, skiprows=4)

# Preview the data
df.head()
```

```python

# List all column names
df.columns
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Choose the latest year
latest_year = '2023'

# Get top 10 countries by population
top10 = df[['Country Name', latest_year]].dropna().sort_values(by=latest_year, ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x=latest_year, y='Country Name', data=top10, palette='mako')
plt.title(f'Top 10 Most Populous Countries in {latest_year}')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig("top10_population_2023.png")  # Saves image
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.histplot(df[latest_year].dropna(), bins=30, kde=True, color='skyblue')
plt.title(f'Population Distribution of All Countries in {latest_year}')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.savefig("population_distribution_2023.png")
plt.show()
```
